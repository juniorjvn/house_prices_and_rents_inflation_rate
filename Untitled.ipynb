```python
home = pd.read_csv('data.csv')
home
```

```python
home['BUILDING CLASS CATEGORY'].unique()
```

```python
home['BUILDING CLASS CATEGORY']  = home['BUILDING CLASS CATEGORY'].str.slice(start=2)
home['BUILDING CLASS CATEGORY']  = home['BUILDING CLASS CATEGORY'].str.strip()
```

```python
home['SALE DATE'] = home['SALE DATE'].str.slice(stop=4)
home = home[home['BOROUGH'] == 4]
home = home[home['SALE PRICE'] > 100000]
home = home.sort_values('SALE DATE')
home
# home = home[(home['BUILDING CLASS CATEGORY'] == 'ONE FAMILY HOMES') | (home['BUILDING CLASS CATEGORY'] == 'TWO FAMILY HOMES') |
#            (home['BUILDING CLASS CATEGORY'] == 'THREE FAMILY HOMES')]

home = home[(home['BUILDING CLASS CATEGORY'] == 'ONE FAMILY DWELLINGS') | (home['BUILDING CLASS CATEGORY'] == 'TWO FAMILY DWELLINGS') |
           (home['BUILDING CLASS CATEGORY'] == 'THREE FAMILY DWELLINGS')]

# home = home[home['BUILDING CLASS CATEGORY'] == 'THREE FAMILY HOMES']
```

```python
grp = home.groupby(['SALE DATE']).agg({'SALE PRICE': np.mean})
grp
```

```python
q_zip = home[home['BOROUGH'] == 4]['ZIP CODE'].dropna().unique()
```
